# ***18. Count Number of Action Movies***

```python
len(data[data['Genre'].str.contains('action',case=False)])
```

```python
all_genres = data['Genre'].str.split(',').apply(lambda x: [genre.strip() for genre in x])
flat_genres = [genre for sublist in all_genres for genre in sublist]
genre_count = pd.Series(flat_genres).value_counts()
genre_count
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_count.index, y=genre_count.values, palette='Blues_d')
plt.xticks(rotation=45)
plt.title("Number of Films Made in Each Genre", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Films", fontsize=12)
plt.tight_layout()
plt.show()
```
